```python
#drop column Alley as it is having more null value

dataset1.drop(['Alley'],axis=1, inplace=True)
dataset1
```

```python
# checking null value in MasVnrType column

#dataset1['MasVnrType'].isnull().sum()
dataset1['MasVnrType'].isna().sum()
```

```python
# checking datatype of MasVnrType column

dataset1['MasVnrType'].dtypes
```

```python
dataset1['MasVnrType'].unique()
```

```python
#replace MasVnrType missing values with mode

masvnr_object = [col for col in dataset1 if dataset1['MasVnrType'].dtype=='O']

for masvnrtype in masvnr_object:
    dataset1['MasVnrType'] = dataset1['MasVnrType'].fillna(dataset1['MasVnrType'].mode()[0])
```

```python
dataset1['MasVnrType'].isna().sum()
```

```python
# checking null value in GarageYrBlt column

dataset1['GarageYrBlt'].isna().sum()
```

```python
# checking datatype of GarageYrBlt column

dataset1['GarageYrBlt'].dtypes
```

```python
dataset1['GarageYrBlt'].unique()
```

```python
# converting 'GarageYrBlt' from float to int
#dataset1 = dataset1.astype({"GarageYrBlt": 'int'})

dataset1['GarageYrBlt'] = dataset1['GarageYrBlt'].replace(to_replace=np.nan, value = 0).astype('int64')
```

```python
dataset1['GarageYrBlt'].dtypes
```

```python
#dataset1['GarageYrBlt'].mean()

dataset1['GarageYrBlt'] = dataset1['GarageYrBlt'].fillna(
    dataset1['GarageYrBlt'].mean()
)
dataset1['GarageYrBlt']
```

```python
dataset1['GarageYrBlt'].isna().sum()
```

```python
# checking null value in Electrical column

dataset1['Electrical'].isna().sum()
```

```python
#dataset1['Electrical'].mode()
dataset1['Electrical'] = dataset1['Electrical'].fillna(
    dataset1['Electrical'].mode()
)

#replace Electrical missing values with mode

elec_object = [col for col in dataset1 if dataset1['Electrical'].dtype=='O']

for masvnrtype in elec_object:
    dataset1['Electrical']  = dataset1['Electrical'] .fillna(dataset1['Electrical'] .mode()[0])
```

# **Univariate Analysis**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Kernel density plot of house prices
sns.kdeplot(dataset1['SalePrice'])
plt.xlabel('House Price')
plt.ylabel('Density')
plt.title('Kernel Density Plot of House Prices')
plt.show()
```

```python
# Histogram of house prices
plt.hist(dataset1['SalePrice'], bins=50)
plt.xlabel('House Price')
plt.ylabel('Frequency')
plt.title('Distribution of House Prices')
plt.show()
```

```python

plt.figure(figsize=(12,6))
sns.histplot(dataset1['LotArea'], kde=True, bins=30)
plt.title('Distribution of Houses by Lot Size')
plt.show()
```

```python
# Boxplots can help identify outliers
plt.figure(figsize=(8, 4))
sns.boxplot(x=dataset1['SalePrice'])
plt.title(f'Boxplot of Sale Price')
plt.show()
```
