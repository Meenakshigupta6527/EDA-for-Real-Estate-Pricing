```python
df_num = dataset1.select_dtypes(include=['number'])  # Select numeric columns
df_num
```

```python
df_num.hist(figsize=(20,20), bins=50, xlabelsize=8, ylabelsize=8);
```

```python
df_cat_encoded = pd.get_dummies(dataset1_cat, drop_first=True)  # Apply one-hot encoding
df_cat_encoded
```

```python
# Identify boolean columns and convert them to numeric
boolean_cols = df_cat_encoded.columns[df_cat_encoded.dtypes == 'bool']
df_cat_encoded[boolean_cols] = df_cat_encoded[boolean_cols].astype(int)
```

```python
# Concatenate the numerical and encoded categorical columns
df_ml = pd.concat([df_num, df_cat_encoded], axis=1)  # Concatenate along columns (axis=1)
df_ml
```
