```python
#dataset1_cat = dataset1.select_dtypes(np.object)
dataset1_cat = dataset1.select_dtypes(include ='object')
dataset1_cat
```

```python
df_num = dataset1.select_dtypes(include=['number'])  # Select numeric columns
df_num
```

```python
df_cat_encoded = pd.get_dummies(dataset1_cat, drop_first=True)  # Apply one-hot encoding
df_cat_encoded
```

```python
# Identify boolean columns and convert them to numeric
boolean_cols = df_cat_encoded.columns[df_cat_encoded.dtypes == 'bool']
df_cat_encoded[boolean_cols] = df_cat_encoded[boolean_cols].astype(int)
```

```python
# Concatenate the numerical and encoded categorical columns
df_ml = pd.concat([df_num, df_cat_encoded], axis=1)  # Concatenate along columns (axis=1)
df_ml
```

# **Univariate Analysis**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.histplot(df_ml['SalePrice'], kde=True, bins=10)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.histplot(df_ml['LotArea'], kde=True, bins=30)
plt.title('Distribution of Houses by Lot Size')
plt.show()
```

```python
# Boxplots can help identify outliers
plt.figure(figsize=(8, 4))
sns.boxplot(x=df_ml['SalePrice'])
plt.title(f'Boxplot of Sale Price')
plt.show()
```
